```python
Q1. What is multiprocessing in python? Why is it useful?
Answer--Multiprocessing in Python refers to the ability of a 
program to utilize multiple processes to achieve parallelism and perform tasks concurrently. 
Multithreading is useful for several reasons:
1--Parallelism
2--Avoiding GIL Limitattions
3--Resource isolation
4-Improved performance
5--Fault isolation
6--Scalability
7--Use of external libraries

Q2. What are the differences between multiprocessing and multithreading?
Answer--Multiprocessing and multithreading are both techniques used to achieve concurrent execution
in programs,
but they have distinct differences in terms of how they utilize resources and manage concurrency.
a--Process vs threads 
b--Resource utillization
c--communication
d--isolation and fault tolerance 
e--memory consumption
f--Startup overhead

Q3. Write a python code to create a process using the multiprocessing module.
Answer--import multiprocessing

def worker_function(name):
    print(f"Worker {name} is executing.")

if __name__ == "__main__":
    # Create a Process object
    process = multiprocessing.Process(target=worker_function, args=("Process 1",))

    # Start the process
    process.start()

    # Wait for the process to finish
    process.join()

    print("Main process has finished.")
Q4. What is a multiprocessing pool in python? Why is it used?
Answer--A multiprocessing pool in Python is a convenient way to distribute the execution of 
multiple function calls across multiple processes. It provides a high-level interface for parallel
execution,
allowing you to perform tasks concurrently by spreading them across a pool of worker processes
Advantages of using a multiprocessing pool:
1.Efficient resource Utillizatation
2.Simplified API
3-Task distribution
4.Automatic process management
5-Optimized for input output tasks

Q5. How can we create a pool of worker processes in python using the multiprocessing module?
Answer--Creating a pool of worker processes in Python using the multiprocessing module is 
straightforward. The Pool class from the multiprocessing module allows you to create 
a pool of worker processes and provides methods to distribute tasks among those processes.
import multiprocessing

def worker_function(number):
    return number * 2

if __name__ == "__main__":
    numbers = [1, 2, 3, 4, 5]

    # Create a pool of worker processes
    with multiprocessing.Pool(processes=2) as pool:
        # Distribute tasks using the map function
        results = pool.map(worker_function, numbers)

    print("Results:", results)
Q6. Write a python program to create 4 processes, each process should print a different
number using the
multiprocessing module in python.
Answer--
```
